```python
#assignment no. 9

#1
 import simpy
import random

waiting_time = []

def customer(env, name, server):
    arrival = env.now
    print(name, "arrived at", arrival)

    with server.request() as req:
        yield req

        wait = env.now - arrival
        waiting_time.append(wait)

        print(name, "started service at", env.now)

        service_time = random.randint(2, 5)
        yield env.timeout(service_time)

        print(name, "left at", env.now)

def source(env, server):
    for i in range(5):
        yield env.timeout(random.randint(1, 3))
        env.process(customer(env, "Customer " + str(i + 1), server))

env = simpy.Environment()
server = simpy.Resource(env, capacity=1)

env.process(source(env, server))
env.run()

print("\nAverage Waiting Time =", sum(waiting_time) / len(waiting_time))


#2
  server = simpy.Resource(env, capacity=2)


#3
 service_time = random.randint(4, 6)

#4
 print("Average Waiting Time =", sum(waiting_time) / len(waiting_time))
```
