```python
#assignment no. 8

#1 Python script to sort a dictionary by value(ascending and descending)
d = {'a': 5, 'b': 2, 'c': 8, 'd': 1}
# Ascending order
asc = dict(sorted(d.items(), key=lambda item: item[1]))
print("Ascending Order:", asc)
# Descending order
desc = dict(sorted(d.items(), key=lambda item: item[1], reverse=True))
print("Descending Order:", desc)


#2 Python script to check if a given key already exists in a dictionary
d = {'name': 'Sujal', 'age': 18}
key = input("Enter key to check: ")
if key in d:
    print("Key exists")
else:
    print("Key does not exist")


#3 Python script to merge two python dictionary
d1 = {'a': 1, 'b': 2}
d2 = {'c': 3, 'd': 4}
d3 = {**d1, **d2}
print("Merged Dictionary:",d3)


#4 Python program to add an item in a tuple
t = (1, 2, 3)
   # Adding item using concatenation
t = t + (4,)
print("Updated Tuple:", t)

#5 Python program to create a tuple with different data types
t = ("Sujal", 18, 75.5, True)
print("Tuple:", t)

#6 Python program to sum all the items in a list
lst = [10, 20, 30, 40]
total = sum(lst)
print("Sum of list items:", total)

#7 Python program to get the largest no. from a list
lst = [12, 45, 7, 89, 34]
largest = max(lst)
print("Largest Number:", largest)

#8 Python program to add members in a set
s = {1, 2, 3}
  # Add single item
s.add(4)
  # Add multiple items
s.update([5, 6])
print("Updated Set:", s)

#9 Python program to reverse the order of the items in the array
from array import *
arr = array('i', [1, 2, 3, 4, 5])
arr.reverse()
print("Reversed Array:", arr)

#10 Python program to create an array of 5 integers and display the array items.
from array import *
arr = array('i', [10, 20, 30, 40, 50])
print("Array Elements:")
for i in arr:
    print(i)
  # Accessing elements using index
print("First Element:", arr[0])
print("Second Element:", arr[1])
print("Third Element:", arr[2])
```
